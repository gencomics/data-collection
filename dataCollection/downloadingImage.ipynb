```python
baseUrl = 'http://pt.jikos.cz'
```

```python
## imageLinks
imageLinks = dict()
```

```python
yearLinks = dict()
for aTag in aTags:
    if len(aTag.text) == 4:
        yearLinks[aTag.text] = baseUrl+aTag['href']
yearLinks
```

```python
MONTH_DICT = {
    'January':'1',
    'February':'2',
    'March':'3',
    'April':'4',
    'May':'5',
    'June':'6',
    'July':'7',
    'August':'8',
    'September':'9',
    'October':'10',
    'November':'11',
    'December':'12'
}
```

```python
from PIL import Image
from io import BytesIO
import os

def initDir(path):
    try:
        os.makedirs(path)
    except OSError as exc: # Python >2.5
        if exc.errno == errno.EEXIST and os.path.isdir(path):
            pass
        else: raise


def saveImageToFile(imageLink, path):
    response = get(imageLink)
    if response.status_code == 200:
#         print(response)
        i = Image.open(BytesIO(response.content))
        names = imageLink.split("/")
        filename = names[len(names)-1]
        i.save(path+filename, "gif")
        print(path+filename)
#         
```

```python
os.getcwd()
```

```python
afterYear = 2013

monthLinks = dict()
for year in yearLinks:
    if int(year) <= afterYear:
        continue
    yearUri = yearLinks[year]
    response = get(yearUri)
    year_soup = bs(response.text, 'html.parser')
    #print (type(year_soup.p))
    ps = year_soup.findAll('p')
    months = ps[1].text.split("\n") # 2nd <p> element : holds month urls
    months = [month for month in months if len(month)>0]
    for month in months:
        relPath = os.getcwd()+"/garfield/"+year+"/"+month+"/"
        initDir(relPath) # making dir if dosent exist
        
        monthUri = yearUri+MONTH_DICT[month]+"/"
        response = get(monthUri)
        month_soup = bs(response.text, 'html.parser')
        imageTags = month_soup.findAll('img')
        #print(imagesLink)
        for imageTag in imageTags:
            imageLink = imageTag['src']
            if 'http' in imageLink:
                saveImageToFile(imageLink, relPath) # saving to file in relPath
        
        
        
        
    
```
