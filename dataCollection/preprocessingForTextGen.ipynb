```python
import pandas as pd
```

```python
strList = list()
f = open('garfield.txt', 'r', encoding = "ISO-8859-1")
fl = f.readlines()
for l in fl:
    print(l[12:])
    strList.append(l[12:])
```

```python
print(strList)
```

```python
with open('garfield_text_gen.txt', 'w') as filehandle:
    filehandle.writelines("%s\n" % line for line in strList)
```
